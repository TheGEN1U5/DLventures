```python
import numpy as np
```

```python
embed_file = "glove.6B/glove.6B.300d.txt"
embeds = {}

with open(embed_file, "r", encoding='utf-8') as f:
    
    for l in f:
        l_split = l.split()
        word = l_split[0]
        vector = np.asarray(l_split[1:], "float32")
        embeds[word] = vector
```

```python
import scipy.spatial as spatial
```

```python
def find_closest_embeddings(embedding):
    return sorted(embeds.keys(), key=lambda word: spatial.distance.cosine(embeds[word], embedding))
```

```python
find_closest_embeddings(embeds["fart"])[:10]
```
